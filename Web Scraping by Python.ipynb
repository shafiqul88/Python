```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://dsebd.org/dse30_share.php'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
table = soup.find('table', class_='table-bordered')
```

```python
table_head = table.find_all('th')
```

```python
table_head_title =[title.text.strip() for title in table_head]
```

```python
print(table_head_title)
```

```python
import pandas as pd
df = pd.DataFrame(columns = table_head_title)
```

```python
df
```

```python
row_data = table.find_all('tr')
```

```python
for row in row_data[1:]:
    myrow_data = row.find_all('td')
    each_row_data = [data.text.strip() for data in myrow_data]
    
    length = len(df)
    df.loc[length] = each_row_data
```
